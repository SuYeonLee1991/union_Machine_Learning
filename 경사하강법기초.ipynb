```python
def get_weight_updates (w1, w0, X, y, learning_rate=0.01 ):
    N = len(y)
    # w1_update, w0_update  wl, wo shape    0  
    wl_update = np. zeros_like(w1)
    w0_update = np. zeros_like(w0)
    #       
    y_pred = np.dot (X, w1.T) + w0
    diff = y-y_pred

    
    
    #wo_update dot      1    
    w0_factors = np.ones((N,1))
    #w1 WO  w1_update w0_update 
    w1_update = -(2/N) * learning_rate* (np.dot (X.T, diff))
    w0_update = -(2/N) * learning_rate* (np.dot (w0_factors.T, diff))
   
    return w1_update, w0_update, diff
```

```python
#  iters    w1 WO  . 
def gradient_descent_steps (X, y, iters=10000):
    epoch=0
    #wO w1  0 .
    w0= np.zeros((1,1))
    w1 = np.zeros((1,1))
    N = len(y)
    #   iters   get_weight_updates()  wl, wo  .
    for ind in range (iters):
        w1_update, w0_update ,diff  = get_weight_updates (w1, w0, X, y, learning_rate=0.01)
        w1= w1 - w1_update
        w0= w0 - w0_update
        epoch  += 1
        if epoch % 10 == 0:
            
            mse = np.sum(np.square (diff))/N #
            print(f'epoch : {epoch}  mse : {mse}, w1 : {w1}, w0 : {w0}')
    return w1, w0
```
